```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn import datasets
from sklearn import manifold
%matplotlib inline
```

```python
train = pd.read_csv('data/train_mnist.csv')
X = train.drop('label', axis=1)
y = train['label']
```

```python
X.shape, y[0]
```

```python
single_image = X.loc[0].values.reshape(28,28)
plt.imshow(single_image, cmap='gray')
```

```python
tsne = manifold.TSNE(n_components=2, random_state=42)
transformed_data = tsne.fit_transform(X[:3000].values)
```

```python
tsne_df = pd.DataFrame(np.column_stack((transformed_data, y[:3000].values)), columns=['x','y','targets'])
tsne_df.targets = tsne_df.targets.astype(int)
```

```python
grid = sns.FacetGrid(tsne_df, hue='targets', size=8)
grid.map(plt.scatter, "x", "y").add_legend()
```

```python
wine = pd.read_csv('data/winequality-red.csv')
```

```python
wine.quality.value_counts()
```

```python
map_quality = {3:0, 4:1, 5:2, 6:3, 7:4, 8:5}
wine.quality = wine.quality.map(map_quality)
```

```python
wine.quality.value_counts()
```

```python
df = wine.sample(frac=1).reset_index(drop=True)
df.shape
```

```python
train = df.head(1000)
test = df.tail(599)
```

```python
from sklearn import tree
from sklearn import metrics

clf = tree.DecisionTreeClassifier(max_depth=3)
clf.fit(train.drop('quality', axis=1), train.quality)
```

```python
train_preds = clf.predict(train.drop('quality', axis=1))
test_preds = clf.predict(test.drop('quality', axis=1))

train_accuracy = metrics.accuracy_score(train.quality, train_preds)
test_accuracy = metrics.accuracy_score(test.quality, test_preds)

print('train accuracy: {}\n test accuracy: {}'.format(train_accuracy, test_accuracy))
```

```python
import matplotlib
matplotlib.rc('xtick', labelsize=20)
matplotlib.rc('ytick', labelsize=20)

train_accuracies = [0.5]
test_accuracies = [0.5]
```

```python
for i in range(1, 25):
    clf = tree.DecisionTreeClassifier(max_depth=i)
    clf.fit(train.drop('quality', axis=1), train.quality)
    
    train_preds = clf.predict(train.drop('quality', axis=1))
    test_preds = clf.predict(test.drop('quality', axis=1))

    train_accuracy = metrics.accuracy_score(train.quality, train_preds)
    test_accuracy = metrics.accuracy_score(test.quality, test_preds)
    
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)
```

```python
plt.figure(figsize=(10,5))
sns.set_style('whitegrid')
plt.plot(train_accuracies, label="train accuracies")
plt.plot(test_accuracies, label="test accuracies")
plt.legend(loc="upper left", prop={'size':15})
plt.xticks(range(0,26,5))
plt.xlabel("max_depth", size=20)
plt.ylabel("accuracy", size=20)
plt.show()
```

These types of models where training accuracy is high and testing accuracy is low are known to be **Overfitting**
